```python
# Construct an example graph 
# Note that this is the same graph that is returned by:
# G = create_example_graph()

G = EmbeddedGraph()

G.add_node('A', 1, 2)
G.add_node('B', 3, 4)
G.add_node('C', 5, 7)
G.add_node('D', 3, 6)
G.add_node('E', 4, 3)
G.add_node('F', 4, 5)

G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'E')
G.add_edge('C', 'D')
G.add_edge('E', 'F')

G.set_centered_coordinates()

G.plot()
plt.savefig('paper/figures/example_graph.png', bbox_inches='tight', dpi=300)
```

```python
G.plot(color_nodes_theta=7*np.pi/4)
plt.title('Coloring nodes with $g_\omega$ for  $\\theta = 7\pi/4$')
plt.savefig('paper/figures/example_graph_7pi_over_4.png', bbox_inches='tight', dpi=300)
```

```python
G.plot(color_nodes_theta=2*np.pi/4)
plt.title('Coloring nodes with $g_\omega$ for  $\\theta = \pi/2$')
plt.savefig('../paper/figures/example_graph_pi_over_2.png', bbox_inches='tight', dpi=300)
```

```python
myect = ECT(num_dirs = 100, num_thresh=100)
myect.set_bounding_radius(1.2 * G.get_bounding_radius())

myect.calculateECC(G, np.pi/2,bound_radius=True)
```

```python
myect.plotECC(G, -np.pi/2)
```

```python
myect.calculateECT(G)

M = myect.get_ECT()

# We can use the built in command to plot the matrix. Unlike the plotECC function, this command does not calculate the ECT when called so it must have been run earlier. An equivalent command is myect.plotECT()
myect.plot('ECT')

plt.savefig('paper/figures/example_ect.png', bbox_inches='tight', dpi=300)
```
